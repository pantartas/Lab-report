There are two applicable assumptions that will be described separately: \textit{inviscid flow} and \textit{incompressible flow}. These two assumptions will greatly simplify our theoretical model.

#### Inviscid flow

There are no truly inviscid flows in nature as the phenomena of friction (viscosity), mass diffusion and thermal conduction always occur at a molecular level to a certain degree. However, in many engineering applications, these transport phenomena are mostly negligible, especially for high and finite Reynolds numbers flows. Therefore, the flow can be separated into two regions: the inviscid region and the viscous region. In the case of an aerofoil mounted in a flow at a low angle of attack, the viscous region is small and represented by the thin viscous boundary layer neighbouring the surface of the aerofoil. The inviscid region is represented by all the flow outside the boundary layer. **Given a wing chord $c=0.063m$, an upstream flow velocity $u=20 m/s$ and a kinematic viscosity $\nu= 15.11\times10^{-6} m^{2}/s$ ($Re=\frac{u\times c}{\nu}$), how well is this assumption holding? Also, can the flow still be considered inviscid when the aerofoil is at high angles of attack at which separation occurs and hence a large wake is formed downstream?}**

#### Incompressible flow

An incompressible flow is defined as a flow for which the working fluid density $\rho$ is constant. There are, of course, more advanced definitions of what compressibility is which link to thermodynamic considerations, but a good approximation is that all flows for which Mach number (M) is less than 0.3, can be considered incompressible. **Hence, what are your comments regarding the applicability of the incompressibility assumption for the current case?**

#### Bernoulli's Equation
Using the assumptions of the inviscid and the incompressible flow, Bernoulli's equation becomes applicable. The equation is: 

\begin{equation}
p+\frac{1}{2}\rho u^2=const.
\label{Eqn:bernoulli1}
\end{equation}

<h5><center>Equation 1</center></h5>

Thus for an incompressible, inviscid flow, the static and dynamic pressures add up to a constant along streamlines.

#### Irrotational flow
If we consider an infinitesimal fluid element moving through the flow field, we can describe its orientation as a combination of both translation and rotation. Additionally, the fluid element may also change its shape, but this change can also be considered a combination of translation and rotation. The case on the right of Figure 1 shows on the upper streamline a fluid element where the angular velocities of its sides are zero, hence the flow is irrotational. The lower streamline of the case on the right shows a two dimensional fluid element where the angular velocities of the two intersecting sides are finite, but equal and opposite to each other and so their sum is identically zero; hence the flow is irrotational as well. **How well is this assumption holding for the current case? If besides inviscid and incompressible, the flow is also irrotational, the value of the constant from Equation 1 is the same between all streamlines. Applying Bernoulli's equation far upstream:

\begin{equation}
p_{\infty}+\frac{1}{2}\rho u_{\infty}^2=p_{0}

<h5><center>Equation 2</center></h5>

where $p_{0}$ is defined as the stagnation or total pressure and the subscript $\infty$ refers to the upstream conditions.

#### Pressure coefficient
The static pressures and velocities around the aerofoil differ from their upstream values. Assuming that there is a negligible variation of static pressure through the thin boundary layers on the surface of the aerofoil, Bernoulli's equation can be applied to relate the surface static pressure $p_{s}$ to the upstream conditions:

\begin{equation}
p_{0}=p_{s}+\frac{1}{2}\rho u^2=p_{\infty}+\frac{1}{2}\rho u_{\infty}^2
\label{Eqn:bernoulli3}
\end{equation}
<h5><center>Equation 3</center></h5>

where $u$ is the velocity just outside the boundary layer. By defining the surface pressure coefficient, $C_{p}$ in the usual way, from Equation 3:

\begin{equation}
C_{p}=\frac{p_{s}-p_{atm}}{\frac{1}{2}\rho u_{\infty}^2}=\frac{p_{s}-p_{atm}}{p_{0}-p_{\infty}}
\label{Eqn:pressurecoeff}
\end{equation}
<h4><center>Equation 4</center></h4>

For the airfoil, we will assume the static pressure farther away from the airfoil surface is $P_{atm}$ and the velocity is calculated from the tunnel free stream pressure and total pressure. This will be measured through the use of a manometer, where the pressure is proportional to the column height and the inclination angle of the manometer (FEEG1003 Thermofluids for revision on how to use a manometer). 

The experimental setup is fitted with a number of pressure tappings on both the top and bottom surfaces that are connected to a multi-manometer; this will allow you to measure the static pressures at points around the aerofoil. A standard pitot-static probe is mounted at the inlet to the contraction and the inlet of the test section to read the stagnation pressure as well as the freestream static pressure. The coefficients of pressure around the aerofoil can be determined using these readings.

Once you have the pressure coefficents you can integrate them to calcuate lift and drag. Note that the drag you will measure is only the pressure drag and would not include the viscous contribution.

In general terms an aerofoil at a geometric angle of attack $ \alpha$.

The objective now is to determine the lift coefficients using the readings provided by the pressure tappings. First, consider an element of unit span and length $\Delta s$ on the surface of the aerofoil (i.e. left small purple line) which is tilted by $\theta$ with respect to the aerofoil's axis (i.e. red line). The normal net force to this element is, under the aforementioned assumptions, $p_{s}\Delta s$ which in the XY Cartesian coordinate system defined by the aerofoil has the following components:

\begin{equation}
\begin{split}
& \Delta N_{a}=\pm p_{s}\cos(\theta)\Delta s=\pm p_{s}\Delta x \\
& \Delta T_{a}=\pm p_{s}\sin(\theta)\Delta s=\pm p_{s}\Delta y
\label{Eqn:force1}
\end{split}
\end{equation}
<h4><center>Equation 5</center></h4>

The $+$ elements represent the ones which create a force component in positive Y or X direction and $-$ elements represent the ones that create a force component in negative Y or X direction.

The value of the normal force (N) is the integration of all $\Delta N$ across the surface of the entire airfoil. Since in the current case the size of $\Delta s$ is not infinitesimal and has a finite determinable value, the integration becomes a summation of the normal force across the upper and lower surfaces of the airfoil. The same applies to the tangential force. Hence:
 \begin{equation}
  \begin{split}
   N &=\sum_{up} \Delta N_{a}+\sum_{up} \Delta N_{f}+\sum_{low} \Delta N_{a}+\sum_{low} \Delta N_{f}\\
   T &=\sum_{up} \Delta T_{a}+\sum_{up} \Delta T_{f}+\sum_{low} \Delta T_{a}+\sum_{low} \Delta T_{f}
  \label{Eqn:force3}
  \end{split}
 \end{equation}
 <h4><center>Equation 6</center></h4>

 where N is the normal force per unit span and T is the tangential force per unit span, both relative to the XY Coordinate system belonging to the aerofoil. The above summation can be re-written as an integral and converted to a force coeffcient as follows: 
 
  \begin{equation}
   \begin{split}
    C_{N} &=\frac{N}{\frac{1}{2}\rho u_{\infty}^2 c}=\frac{1}{c} \cdot \int_{0}^{c} (C_{p,l} - C_{p,u}) \cdot dx\\ 
    C_{T} &=\frac{T}{\frac{1}{2}\rho u_{\infty}^2 c}=\frac{1}{c} \cdot \int_{0}^{c} (C_{p,u} \cdot \frac{dy_{u}}{dx} - C_{p,l} \cdot \frac{dy_{l}}{dx}) \cdot dx
   \label{Eqn:force4}
   \end{split}
  \end{equation}
<h4><center>Equation 7</center></h4>

Here $\rho$ is the air density, $u_{\infty}$ is the upstream wind velocity and $c$ is the wing chord. $\frac{dy_{u}}{dx}$ and $\frac{dy_{l}}{dx}$ are slope of the aerofoil surface at the location of the tappings on the upper and lower surfaces. 

These normal and tangential forces can be convered to lift and drag coefficients. Note that for very thin airfoils, the pressure drag is basically zero as the slope of the airfoil is very small.Regardless, we can determine the the coefficient of lift that can be determined as,

 \begin{equation}
      C_{l} =C_{N}\cos(\alpha)-C_{T}\sin(\alpha)
    \label{Eqn:force5}
    \end{equation}
    
   <h4><center>Equation 8</center></h4>

For the above equation, if $\alpha$ is small, then, the normal force and the lift force will be very similar to each other. 

### Thin aerofoil theory
The objective of this brief section is to refresh assumptions under which this theory is applicable and also, to refresh some of its consequences. Requirements:
* As the name states, the maximum aerofoil thickness must be small compared to the chord length. You will be using a NACA0020 airfoil section where the maximum thickness of the wing is approximately 12.6 $mm$ and the chord is 63 $mm$.
* The theory is restricted to low angles of attack.
* The flow must be both inviscid and incompressible.
* The flow must be two-dimensional.

**Given the current experimental conditions, how applicable are these assumptions?**
This theory predicts the following key points for aerofoils which respect the aforementioned requirements:
* Lift coefficient is directly proportional with AoA and the proportionality parameter is $2\pi$ (i.e. $\frac{dC_{L}}{d\alpha}=2\pi$).
* On a symmetric aerofoil - no camber - the centre of pressure is located at the quarter-chord behind LE.
* For any aerofoil - with or without camber - the aerodynamic centre is located at the quarter-chord behind LE.

Numerically, the coefficient of lift of an aerofoil within the assumptions of thin aerofoil theory is,

  \begin{equation}
       C_{l}==2\pi(\alpha-\alpha_{0})
      \label{Eqn:thinaerofoil1}
     \end{equation}
<h4><center>Equation 9</center></h4>

where $\alpha_{0}$ represents the angle of attack at zero lift and $\alpha$ represents the \textit{geometric} AoA. The value $2\pi$ is the theoretical lift curve slope and in your experiment coudl give an answer slighty different than this. The value of $\alpha_0$ represents the added effect of the camber on the coefficient of lift. For a symmetric aerofoil, the value of $\alpha_0$ is 0. Furthermore, the value of $\alpha_0$ can be either positive or negative, depending on a positive or negative camber, respectively. The first step is to define a function to find the velocity of the flow taking the total and static pressure as inputs.

Following this define a function to calculate the pressure coefficent taking the velocity, static pressure at the tapping and the atmospheric pressure.

Once the two above functions have been define, create a function to take the array for each angle of attack as set out above, The first 3 values can be used to calculate the velocity the flow, while the others can be used to calculate the Cp on the upper and lower surfaces. Remember to use the x coordinate array that we made earlier for the lower and upper surfaces. These values in terms of $x/c$ is also given in the file. Split the pressures in to two lists/arrays for the lower and upper surfaces. This will make it easy to carry out the integration and also useful for plotting. 

The slope of the aerofoil will be needed and this can be calculated from the aerofoil equation for an aerofoil, the tapping locations can be found above on the plot and therefore the gradient can be calculated. 

$y_t = 5 \cdot t \cdot [0.2969 \cdot \sqrt{x} - 0.1260 \cdot x -0.3516 \cdot x^2 - 0.1015 \cdot x^4]$

You can just compute the derivative analytically and then use that formula to get $dy_u/dx$ and $dy_l/dx$. This way, you can just get the values at the location of the pressure tapping. 

Note that in this equation all units are non-dimensionalised with respect to chord length (so x = 1 is the trailing edge of the airfoil and t is the thickness of the airfoil in terms of chord length). In order to find $C_N$ and $C_T$, you will need to integrate the pressure distribution. This can be done by using the trapezium rule or the simpson's rule. 

The arrays to be used in the data processing are defined in the summary and load data cell, the arrays are from Pandas data frames as introduced at the begining of this module. 

```python
# Define a function to calculate the Velocity
def findvelocity(ptot,pstat):
    v = (2*(ptot-pstat)/rho)**0.5
    return v

# Define a function to calculate the Cp
def findcp(v,pstat,patm):
    cp = (pstat-patm)/(0.5*rho*v**2)
    return cp


# Define a function to process the array
def process(array):
    patm = array.iloc[0,1]
    ptot = array.iloc[1,1]
    pstat = array.iloc[2,1]
    LWR = array.iloc[3:9,1]
    UPPER = array.iloc[9:15,1]
    return patm,ptot,pstat,LWR,UPPER
    
# Define a function to find the slope of the aerofoil
def findslope(x, t):
    y = []
    for i in x:
        dy = 5*t*(0.14845*i**(-0.5) - 0.406*i**3 + 0.8529*i**2 - 0.7032*i - 0.126)
        y.append(dy)
    return y


# Define a function function to find Cl by integrating the pressure coefficient
def findcl(cpup,cplow,xup,xlow,dyu,dyl,alfa):
    Cn = np.trapz(cplow,xlow)-np.trapz(cpup,xup)
    Ct = np.trapz(cpup*dyu,xup) - np.trapz(cplow*dyl,xlow)
    Cl = Cn*np.cos(alfa) - Ct*np.sin(alfa)
    return Cl,Cn,Ct
```
