## Finite wing theory

The objective of this lab is to determine the force characteristics of a finite wing with a NACA0012 aerofoil profile and compare these results to the theory.


The coefficient of lift of an finite wing using finite wing theory is known to be, 
  \begin{equation}
       C_{L}= a(\overline{\alpha}-\overline{\alpha_{0}})
      \label{Eqn:lift}
     \end{equation}
<h4><center>Equation 1</center></h4>

where $\overline{\alpha_{0}}$ represents the span-averaged angle of attack at zero lift and $\overline{\alpha}$ represents span-averaged geometric angle of attack (AoA). Note that the angles have to be in radians. The value lift curve slope $a$ is different from $a_0$ of an airfoil section.  The lift curve slope for a finite wing is related to the aspect ratio of the wing and an efficiency factor,

\begin{equation}
       a= \frac{a_0}{1 + \frac{a_0}{e \pi AR}}
      \label{Eqn:slope}
     \end{equation}
<h4><center>Equation 2</center></h4>

where $AR = b^2/S$ is the aspect ratio of the wing with $b$ is the wing span and $S$ is the wing area, $e = 1/(1+\tau)$ is the span efficiency factor. Refer to the lecture notes for further details about the theory. 

It is clear that lift curve slope is different for a finite wing compared to an airfoil and therefore should be determine using the value of $AR$ and $e$. Plotting the lift coefficient against the angle of attack will allow you to determine this slope by fitting a straight line to the data. 

In addition to lift curve slope, the finite end also alter the drag experienced by the wing. In fact, the finite end causes tip vortices that causes a component of the lift to act in the direction of the drag. Therefore, there is an additional drag due to lift contribution in the presence of finite end. This drag due to lift, also known as the induced drag, is typically captured by the induced drag coefficient, $C_{D,i}$

\begin{equation}
       C_{D,i}= \frac{C^2_L}{\pi AR}(1+\delta)
      \label{Eqn:ind}
     \end{equation}
<h4><center>Equation 3</center></h4>

where $\delta$ is the induced drag factor that depends on the distribution of lift across the span. Therefore, the total drag on finite wing is given by, 

\begin{equation}
       D = D_i + D_0
      \label{Eqn:dt}
     \end{equation}
<h4><center>Equation 4</center></h4>

where, $D_0$ is the sum of all the drag other than the induced drag ($D_i$).It is also known as the zero-lift drag, which the drag experienced by the wing at zero lift. This total drag is typically comprised of the skin-friction drag, pressure drag and interference drag. For low angles of attack, the pressure drag and interference drag can be negligible and therefore only the skin-friction (or the viscous) drag is responsible for the $D_0$. 

\begin{equation}
       C_D = C_{D,i} + C_{D,0}
      \label{Eqn:cd}
     \end{equation}
<h4><center>Equation 5</center></h4>

The above equation can be written in terms of lift coefficient using equation 3, 

\begin{equation}
       C_D =  C_{D,0} + kC^2_L
      \label{Eqn:pl}
     \end{equation}
<h4><center>Equation 6</center></h4>

Here, $k = (1+\delta)/\pi AR$. Therefore, there is a quadratic relationship between drag coefficient and lift coefficient. From the above equation it is clear that with just the knowledge of $C_L$ and $C_D$, it is possible to estimate the value of $C_{D,0}$ by finding the intercept of the best fit straight line between $C_D$ and $C^2_L$. Note that that the slope of that line is $k$. 

Finally, it is also possible to determine the efficiency of a wing by computing $C_L/C_D$ (also known as glide ratio). The angle of attack where the efficiency is high is not the same as the angle where lift is high. This is because $C_D$ increases with $C_L$ and hence there is a balance between lift and drag. It is possible to determine the maximum efficiency for a given value of $C_{D,0}$ by finding the maximum of $C_L/C_D$. Here, you need to substitute $C_D = C_{D,0} + kC^2_L$ and differentiate that function with respect to $C_L$ (since the effciency after substitution is only a function of $C_L$). This will give you the value of $C_L$ at which the efficiency can reach a maximum. This together with equation 1 will give you the geometric angle of attack required to reach that value of $C_L$.

The wing is mounted on the stepper motor at the rod that passes through the quarter chord location of the wing. This ensure that the pitching moment on the wing is close to zero (since the quarter chord point is a the location of centre of pressure for a symmetric airfoil). The wing can be set at different angles of attack using the stepper motor and the forces and moments can be recorded from the sensor at each geometric angle of attack. As mentioned before, the lift force is the force in the $Y$ direction, i.e. $L = F_Y$ while the drag force is the along the $X$ direction, $D = F_X$. The units of these forces are in Newtons [N]. ## General Data
The temperature in K and Pressure in Pa are in the below cell. Enter a formula to calculate the Density which will be required to calculate the coefficients. 

```python
T = 298 #Kelvin
P = 101325 #Pascal
R = 287
rho = P/(R*T)#Write the formula to calculate density in kg/m^3 (use R for air as 287 K /Kg K)
```

### Write your code to read the csv file and compute $C_L$, $C_D$ and convert $\alpha$ to radians

```python
Uinf = 20 #m/s
b = 0.725 #m (span)
c = 0.3 #m (chord)
e = 0.95 #You can adjust this value to get best fit to the lift curve slope
AR = b/c

data = pd.read_csv("FWdata.csv")
aa,ab,ac = data.columns # column names
AoA = data[aa]*np.pi/180 # This converts angle of attack to radians
D = data[ab]
L = data[ac]
```

### Use the values of $C_L$to plot the lift curve and extract the lift curve slope by fitting a straight line to the data. Remember to only fit the data to angles up to 10 degrees as above that the flow may have stalled



```python
S = 0.25
V = 20
c = 0.3
cl = L[0:4]/(0.5*rho*V**2*b*c)

plt.plot(AoA[0:4],cl,'x')
plt.ylabel('Lift Coeff.')
plt.xlabel('Angle of Attack (radians)')

m, b0 = np.polyfit(AoA[0:4],cl, 1)

def y(x,b):
    y = m*x+b
    return y

n = y(AoA[0:4],b0)

twoD = pd.read_csv('naca efficiency.csv',skiprows=10)
AA,BB,CC,DD,EE,FF,GG,HH,II,JJ = twoD.columns
alphaxfoil = twoD[AA]*np.pi/180
clxfoil= twoD[BB]


plt.plot(AoA[0:4], n, label = '3D data')
plt.plot(alphaxfoil, clxfoil, label = 'Xfoil data')
plt.axis([-0.01,0.15,-0.1,1])
plt.title('Cl vs angle of attack')
plt.legend(loc='upper left')
plt.savefig('cl vs alpha finite')




a = m
a0 = b0/a

print ('lift curve slope:', a)
```

### Use the values of $C_L$ and $C_D$ to plot the drag polar and use that to estimate the zero lift drag.  

Remember if you plot $C_D$ versus $C^2_L$, then you just need to fit a straight line anf the slope will give you some aerodynamic characteristics and the intercept is the zero-lift drag. See lecture notes!

Assuming the zero lift drag is entirely from viscous effects, compare the zero lift drag coefficient to the skin-friction coefficient. 

```python
cd = D[0:4]/(0.5*rho*V**2*b*c)
cl2 = cl**2
plt.plot(cd,cl2, label='experimental values')
plt.ylabel('Lift Coeff.^2')
plt.xlabel('Drag Coeff.')
plt.title('Drag vs Lift coefficient')

m1, b1 = np.polyfit(cl2,cd, 1)

def z(x):
    z = m1*x+b1
    return z

n1 = z(cl2)

plt.plot(n1, cl2, label ='fit')
plt.legend(loc='upper left')
plt.savefig('cl2 vs cd')
k = m1
cd0 = b1
mu = 1.849e-5
Rec = rho*V*c/mu
cf = 0.074*Rec**-(1/5)/2


print ('zero lift drag:', cd0)
print ('\ncf*S =   ',cf)
print ('cd0*Sw = ',cd0*S)
print ('D/q =    ',D[0]/(0.5*rho*V**2))
```

## Use the values of $C_L$ and $C_D$ to estimate the efficiency and find the angle of attack or lift coefficient for maximum efficiency

```python
glide_ratio = cl/(cd0+k*cl**2)

plt.plot(cl, glide_ratio,'x')
q,w,r=np.polyfit(cl,glide_ratio, 2)
def l(x):
    z = q*x**2+w*x+r
    return z
x = np.linspace(0,0.6,100)
glide = l(x)
plt.plot(x,glide, label='experimental values')

deff2 = np.gradient(glide)

q1,w1,r1 = np.polyfit(x,deff2, 2)
def i(x):
    z = q1*x**2+w1*x+r1
    return z
grad = i(x)
plt.ylabel('Glide ratio')
plt.xlabel('Lift Coeff.')

x_coordinates = [0,0.6]
y_coordinates = [0,0]
plt.plot(x_coordinates, y_coordinates,'--')

target_cl = np.interp(0, deff2, x)-0.15
plt.annotate('target cl',(0.45,16),color='red') 
print ('target cl:',target_cl)

Alpha = (target_cl + a*a0)/a
alphadeg = Alpha * 180/np.pi
print ('\nAngle of attack for max. efficiency: ', Alpha,'rad \n'
       '                                     ', alphadeg, 'deg')

bestglide = target_cl/(cd0+k*target_cl**2)
print ('\nbest glide ratio achievable:   ', bestglide)

plt.plot(target_cl,bestglide,'ro')


twoD = pd.read_csv('naca efficiency.csv',skiprows=10)
AA,BB,CC,DD,EE,FF,GG,HH,II,JJ = twoD.columns
twoDcl = twoD[BB]
twoDcd = twoD[CC]
twoDeff = twoDcl/twoDcd

plt.plot(twoDcl[0:9],twoDeff[0:9], label = '2D theory')

plt.title('efficiency vs cl')
plt.legend(loc='upper left')
plt.savefig('efficiency')
```
